```python
import numpy as np
import pandas as pd
```

```python
def hospital(name):
    
    # 첫줄만 받아서 column으로 
    f = open('제주_동물병원현황.csv', 'r', encoding='cp949')
    data = f.readline()
    a = data.replace('\n', '')  # \n 제거
    column = a.split(',') 
    
    # 전체 다 받아서 d로 
    d = []
    while True:
        data = f.readline()
        a = data.replace('\n', '') 
        b = a.split(',') # split으로 하면 리스트 형태가 된다

        d.append(b) # 리스트 d에 리스트들을 쭉 담아준다
        if data == '':
            break
            
    d.pop(0) # 첫 줄 중복 제거 
    d.pop() # 마지막 줄 '' 제거
    f.close()
    
    d1 = pd.DataFrame(d, columns = column) #pandas 데이터 프레임워크 형태로
    
    isTrue = d1['업소명'] == name # True, False 여부
    result = d1[isTrue] # True 인 것만 
    return result
```
